## 1. 데이터 저장하기



```python
import pandas as pd
```

```python
df = pd.DataFrame({'B':[1,2,3,4], 'C':[4,5,6,8]})
df
```

```python
# 1. CSV로 저장하기
df.to_csv('temp.csv')
```

```python
df.to_csv('./sample_data/temp.csv')
```

```python
# 2. 엑셀 파일로 저장하기
df.to_excel("temp.xlsx")
```

```python
# 3. pickle로 저장하기
df.to_pickle("temp.pkl")
```

```python
# 4. SQLite3 DB로 저장하기
import sqlite3

con = sqlite3.connect("test.db")
df.to_sql("temp", con, if_exists="replace", index=False)
con.close()
```

```python
# 5. html 표로 저장하기
df.to_html('test.html')
```

```python
# 6. json 레이아웃으로 저장하기
df.to_json()
```

```python
df.to_json('test.json')
```
